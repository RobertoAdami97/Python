# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

### Vamos agora enviar um e-mail pelo gmail

#### Use esse código para descobrir qual a posição de um item que queira clicar

- Lembre-se: a posição na sua tela é diferente da posição na minha tela

```python
import pyautogui
import pyperclip
import time
import pandas as pd

pyautogui.PAUSE = 1

# 1- entrar no sistema da empresa
pyautogui.hotkey("ctrl","t")
# o comando abaixo (pyperclip.copy) tem que ser usado, pois o link copiado na mão é zoado, 
# então tem que copiar e colar direto na aba de navegação.
pyperclip.copy(" https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

# 2- navegar no sistema e encontrar a base de vendas
time.sleep(5)
pyautogui.click(x=372, y=314, clicks=2)
# 3- Fazer o download da base de vendas
time.sleep(2)
pyautogui.click(x=415, y=185)
time.sleep(2)
pyautogui.click(x=448, y=580)
time.sleep(5)
# 4- Importar a base de vendas para o python
tabela = pd.read_excel(r"C:\Users\CCE\Downloads\Vendas - Dez.xlsx")
# o paramtro abaixo sheet_name="" é usado quando o excel tem mais de uma aba, ai é so adicionar o nome da aba no sheet_name
# tabela = pd.read_excel(r"C:\Users\CCE\Downloads\Vendas - Dez.xlsx", sheet_name="")
display(tabela)
# 5- Calcular os indicadores da base de vendas
faturamento = tabela["Valor Final"].sum()
print(faturamento)
quantidade = tabela["Quantidade"].sum()
print(quantidade)
```

```python
# 6- Enviar um e-mail para a diretoria com indicadores de venda

# abrir aba
pyautogui.hotkey("ctrl","t")
# entrar no link do email: https://mail.google.com/mail/u/0/#inbox
# o pyperclip.copy é usado pois se o link tiver algum caracter especial, vai dar problema.
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl","v")
pyautogui.press("enter")
time.sleep(5)
# clicar no botão escrever
pyautogui.click(x=114, y=186)
# preencher as informações do e-mail
pyautogui.write("robertoadami@outlook.com")
pyautogui.press("tab") # esse tab seleciona o e-mail
pyautogui.press("tab") # esse tab pula para o campo assunto
pyperclip.copy("Relatório de vendas")
pyautogui.hotkey("ctrl","v")
pyautogui.press("tab") # esse tab pula para o campo corpo do e-mail

# o comando abaixo com tres aspas permite o python escrever um texto com muitas linhas
# o f antes das aspas, é para formatar o texto, com isso, conseguimos colocar as variaveis faturamento e produtos no e-mail
# o comando abaixo (:,.2f) os dois pontos serve para poder usar a formatação nas casas decimais
# A virgula é por que a formatação é em ingles, então a virugla que vai separar as casas decimais
# .2f é para adicionar duas casa depois da virgula
texto = f"""
Prezados,

Segue relatorio de vendas: R${faturamento:,.2f}

Faturamento:
Quantidade de produtos vendidos:{quantidade:,}
Att,

Roberto C Adami
"""
# formatação dos numeros
# esta sendo usado o pyperclip para a formatação, pois o e-mail possui caracteres especiais
pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")
# enviar o e-mail
# ctrl + enter envia o email, pois esse atalho no email faz com que envie ao invez de ter que clicar no botão
pyautogui.hotkey("ctrl","enter")
```
